# Web Scraping using Requests and BeautifulSoup

```python
import requests
from bs4 import BeautifulSoup
```

```python
target_url = "http://phaster.com/zip_code.html"
```

```python
response = requests.get(target_url)
```

```python
response
```

```python
response.status_code
```

```python
soup = BeautifulSoup(response.content, 'lxml')
```

```python
print(soup.prettify())
```

```python
soup.head.title.text
```

```python
tables = soup.find_all('table')
```

```python
len(tables)
```

```python
target_table = tables[0]
```

```python
rows = target_table.find_all('tr')
```

```python
rows[0]
```

```python
len(rows)
```

```python
column, *rows = rows
```

```python
len(rows)
```

```python
rows[0]
```

```python
column
```

```python
fileds = [td.text for td in column.find_all('td')]
```

```python
values = [[td.text for td in row.find_all('td')] for row in rows]
```

```python
[dict(zip(fileds, value)) for value in values]
```
